```python
# tain and test data split
# train_features, train_labels
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels =train_test_split(x,y, test_size = 0.2, random_state = 42)
# X_train, Y_Test, y_train, Y_Test =
```

```python
test_features
```

```python
# scaling the data 
from sklearn.preprocessing import StandardScaler
# Scale data


scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)

```

```python
test_features
```

```python
# Classical Machine Learning: SVM
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Linear
svm_model = SVC(kernel = 'linear', C=1)
svm_model.fit(train_features, train_labels)
```

```python
# Predict
predictions = svm_model.predict(test_features)
```

```python
# accuaracy
accuracy = accuracy_score(test_labels,predictions)
```

```python
print(accuracy)
```

```python
from sklearn.metrics import classification_report
```

```python
# Generate classification report
report = classification_report(test_labels, predictions)

# Print the report for classical SVM Clasifier with linear kernel
print(report)
```

```python
# Classical SVM Model2 with Polynomial kernel
```

```python
# Radial Basis Function kernel (RBF) 

svm_model2 = SVC(kernel = 'rbf', C=1)
svm_model2.fit(train_features, train_labels)
```

```python
# Predict
predictions_rbf = svm_model2.predict(test_features)
```

```python
# accuaracy
accuracy_rbf = accuracy_score(test_labels,predictions_rbf)
```

```python
print(accuracy_rbf)
```

```python
# Generate classification report for Classical SVM with RBF kernal
report_rbf = classification_report(test_labels, predictions_rbf)

# Print the report for Classical SVM with RBF kernal
print(report_rbf)
```

```python
# Classical SVM model 3 with polynomial kernel
```

```python
# Polynomial

from sklearn.svm import SVC

# Define the SVM model using a polynomial kernel
svm_model3 = SVC(kernel='poly', C=1)

# Fit the model on your training data
svm_model3.fit(train_features, train_labels)
```

```python
# Predict
predictions_ploy = svm_model3.predict(test_features)
```

```python
# accuaracy
accuracy_poly = accuracy_score(test_labels,predictions_ploy)
```

```python
print(accuracy_poly)
```

```python
# Generate classification report for Classical SVM with Polynoomial kernel
report_poly = classification_report(test_labels, predictions_ploy)

# Print the report for Classical SVM with Polynoomial kernel
print(report_poly)
```
